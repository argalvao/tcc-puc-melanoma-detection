Carregar e testar as imagens

```python
import os
import torch
from PIL import Image
from torchvision import models, transforms
import matplotlib.pyplot as plt

def load_image(image_path):
    image = Image.open(image_path).convert('RGB')  # Carregar a imagem e converter para RGB
    image_tensor = image_transforms(image)  # Aplicar transformações
    image_tensor = image_tensor.unsqueeze(0)  # Adicionar uma dimensão extra para o batch (1, C, H, W)
    return image, image_tensor  # Retorna a imagem original e a versão transformada
```

Função para realizar a inferência em uma imagem

```python
def predict_image(model, image_path, device):
    model.eval()  
    model.to(device)
    
    image, image_tensor = load_image(image_path)
    image_tensor = image_tensor.to(device)
    
    with torch.no_grad():
        output = model(image_tensor)
        _, prediction = torch.max(output, 1)  # Pega o índice da classe com maior valor
    
    return image_path, image, prediction.item()
```

Função para exibir a imagem junto com a predição

```python
def display_image_with_prediction(image, prediction, image_name):
    plt.imshow(image)
    plt.axis('off')  
    
    if prediction == 1:
        plt.title(f"{image_name}: Melanoma Detectado", color='red')
    else:
        plt.title(f"{image_name}: Sem Melanoma", color='green')
    
    plt.show()
```

Função para realizar a inferência em um diretório

```python
def predict_images_in_directory(model, image_dir, device):
    # Obter todas as imagens do diretório
    for image_name in os.listdir(image_dir):
        image_path = os.path.join(image_dir, image_name)
        if image_path.endswith(('jpg', 'jpeg', 'png')):  # Filtrar por tipos de imagem comuns
            image_path, image, prediction = predict_image(model, image_path, device)
            display_image_with_prediction(image, prediction, image_name)
```

Carregar os modelos

```python
mobilenet = models.mobilenet_v2(pretrained=False)
mobilenet.classifier[1] = torch.nn.Linear(in_features=1280, out_features=2)
mobilenet.load_state_dict(torch.load('/home/abel/Downloads/TCC-PUC-Melanoma-Detection/result_models/mobilenet_v2_melanoma.pth'))

resnet = models.resnet50(pretrained=False)
resnet.fc = torch.nn.Linear(in_features=2048, out_features=2)
resnet.load_state_dict(torch.load('/home/abel/Downloads/TCC-PUC-Melanoma-Detection/result_models/resnet_50_melanoma.pth'))

densenet = models.densenet121(pretrained=False)
densenet.classifier = torch.nn.Linear(in_features=1024, out_features=2)
densenet.load_state_dict(torch.load('/home/abel/Downloads/TCC-PUC-Melanoma-Detection/result_models/densenet121_melanoma.pth'))
```

Testar as imagens

```python
image_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor()
])

image_directory = '/home/abel/Downloads/TCC-PUC-Melanoma-Detection/dataset/model_test'
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# MobileNetV2
print("Resultados MobileNetV2:")
predict_images_in_directory(mobilenet, image_directory, device)

# ResNet-50
print("\nResultados ResNet-50:")
predict_images_in_directory(resnet, image_directory, device)

# DenseNet-121
print("\nResultados DenseNet-121:")
predict_images_in_directory(densenet, image_directory, device)
```
