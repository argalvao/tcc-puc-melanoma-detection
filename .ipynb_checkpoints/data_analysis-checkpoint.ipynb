Análise e Exploração dos Dados (Melanoma)

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

Carregando o dataset unificado (Melanoma)

```python
metadata_path = '/home/abel/Downloads/TCC-PUC-Melanoma-Detection/dataset/metadata.csv'
metadata = pd.read_csv(metadata_path)

# Diretório onde os gráficos serão salvos
output_dir = '/home/abel/Downloads/TCC-PUC-Melanoma-Detection/graficos'
os.makedirs(output_dir, exist_ok=True)

sns.set(style="whitegrid")
```

Relação entre Idade e Diagnóstico (Benigno vs Maligno)

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='benign_malignant', y='age_approx', data=metadata)
plt.title('Relação entre Idade e Diagnóstico (Benigno vs Maligno)')
plt.xlabel('Tipo de Diagnóstico')
plt.ylabel('Idade Aproximada')
plt.savefig(os.path.join(output_dir, 'relacao_idade_diagnostico.png'))
plt.show()
plt.close()
```

Relação entre Sexo e Diagnóstico

```python
plt.figure(figsize=(8,6))
sns.countplot(x='sex', hue='benign_malignant', data=metadata)
plt.title('Relação entre Sexo e Diagnóstico')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.savefig(os.path.join(output_dir, 'relacao_sexo_diagnostico.png'))
plt.show()
plt.close()
```

Distribuição de Diagnósticos por Localização Anatômica

```python
plt.figure(figsize=(10,6))
sns.countplot(y='anatom_site_general', hue='benign_malignant', data=metadata.fillna('Desconhecido'))
plt.title('Distribuição de Diagnósticos por Localização Anatômica')
plt.xlabel('Contagem')
plt.ylabel('Localização Anatômica')
plt.savefig(os.path.join(output_dir, 'diagnosticos_por_localizacao.png'))
plt.show()
plt.close()
```

Relação entre Tipo de Diagnóstico e Localização Anatômica

```python
plt.figure(figsize=(10,6))
sns.countplot(y='anatom_site_general', hue='diagnosis', data=metadata.fillna('Desconhecido'))
plt.title('Relação entre Tipo de Diagnóstico e Localização Anatômica')
plt.xlabel('Contagem')
plt.ylabel('Localização Anatômica')
plt.savefig(os.path.join(output_dir, 'tipo_diagnostico_localizacao.png'))
plt.show()
plt.close()
```

Relação entre Idade e Localização Anatômica

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='anatom_site_general', y='age_approx', data=metadata)
plt.title('Relação entre Idade e Localização Anatômica')
plt.xlabel('Localização Anatômica')
plt.ylabel('Idade Aproximada')
plt.xticks(rotation=45)
plt.savefig(os.path.join(output_dir, 'relacao_idade_localizacao.png'))
plt.show()
plt.close()
```

Relação entre Idade e Espessura do Melanoma

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='age_approx', y='mel_thick_mm', hue='benign_malignant', data=metadata)
plt.title('Relação entre Idade e Espessura do Melanoma')
plt.xlabel('Idade Aproximada')
plt.ylabel('Espessura do Melanoma (mm)')
plt.savefig(os.path.join(output_dir, 'relacao_idade_espessura.png'))
plt.show()
plt.close()
```

Análise e Exploração dos Dados (Não Melanoma)

Carregar do dataset de Não Melanoma

```python
dataset_path = '/home/abel/Downloads/TCC-PUC-Melanoma-Detection/dataset/metadata_2.csv'
dataset_nao_melanoma = pd.read_csv(dataset_path)
```

Distribuição de Idade por Diagnóstico (Benigno vs Indeterminado/Maligno)

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='benign_malignant', y='age_approx', data=dataset_nao_melanoma)
plt.title('Distribuição de Idade por Diagnóstico')
plt.xlabel('Diagnóstico (Benigno/Indeterminado/Maligno)')
plt.ylabel('Idade Aproximada')
plt.savefig(os.path.join(output_dir, 'idade_por_diagnostico_nao_melanoma.png'))
plt.show()
plt.close()
```

Distribuição de Sexo por Diagnóstico

```python
plt.figure(figsize=(8,6))
sns.countplot(x='sex', hue='benign_malignant', data=dataset_nao_melanoma)
plt.title('Distribuição de Sexo por Diagnóstico')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.savefig(os.path.join(output_dir, 'sexo_por_diagnostico_nao_melanoma.png'))
plt.show()
plt.close()
```

Distribuição por Localização Anatômica

```python
plt.figure(figsize=(10,6))
sns.countplot(y='anatom_site_general', hue='benign_malignant', data=dataset_nao_melanoma.fillna('Desconhecido'))
plt.title('Distribuição por Localização Anatômica')
plt.xlabel('Contagem')
plt.ylabel('Localização Anatômica')
plt.savefig(os.path.join(output_dir, 'localizacao_anatomica_nao_melanoma.png'))
plt.show()
plt.close()
```

Relação entre Idade e Localização Anatômica

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='anatom_site_general', y='age_approx', data=dataset_nao_melanoma)
plt.title('Relação entre Idade e Localização Anatômica')
plt.xlabel('Localização Anatômica')
plt.ylabel('Idade Aproximada')
plt.xticks(rotation=45)
plt.savefig(os.path.join(output_dir, 'idade_por_localizacao_anatomica_nao_melanoma.png'))
plt.show()
plt.close()
```

Distribuição de Espessura das Lesões

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='benign_malignant', y='clin_size_long_diam_mm', data=dataset_nao_melanoma)
plt.title('Distribuição de Espessura das Lesões por Diagnóstico')
plt.xlabel('Diagnóstico (Benigno/Indeterminado/Maligno)')
plt.ylabel('Espessura da Lesão (mm)')
plt.savefig(os.path.join(output_dir, 'espessura_lesoes_por_diagnostico_nao_melanoma.png'))
plt.show()
plt.close()
```

Relação entre Idade e Espessura das Lesões
