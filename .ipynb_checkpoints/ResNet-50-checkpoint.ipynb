```python
import torch
from torchvision import models, transforms, datasets
from torch.utils.data import DataLoader
import torch.optim as optim
import torch.nn as nn

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Treinando no dispositivo: {device}")

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Carregar o dataset
train_dataset = datasets.ImageFolder(root='/home/abel/Downloads/TCC-PUC-Melanoma-Detection/dataset/train', transform=transform)
val_dataset = datasets.ImageFolder(root='/home/abel/Downloads/TCC-PUC-Melanoma-Detection/dataset/val', transform=transform)

# Configurar os DataLoaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

Configuração do Modelo

```python
# Carregar o modelo ResNet-50 pré-treinado
model_resnet = models.resnet50(pretrained=True)

# Ajustar para 2 classes
num_ftrs = model_resnet.fc.in_features
model_resnet.fc = torch.nn.Linear(num_ftrs, 2)  # Ajustar para 2 classes

model_resnet = model_resnet.to(device)
```

Configuração do Otimizador e Função de Perda

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_resnet.parameters(), lr=0.001)
```

Treino

```python
num_epochs = 5000
for epoch in range(num_epochs):
    model_resnet.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model_resnet(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * images.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Época {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")

    # Validação
    model_resnet.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model_resnet(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_loss = val_loss / len(val_loader.dataset)
    val_accuracy = correct / total
    print(f"Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}")
```

Salvar o Modelo

```python
torch.save(model_resnet.state_dict(), '/home/abel/Downloads/TCC-PUC-Melanoma-Detection/result_models/resnet_50_melanoma.pth')
```
